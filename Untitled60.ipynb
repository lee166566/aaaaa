```python
import yfinance as yf
import pandas as pd
import numpy as np
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Configuration
ETF_SYMBOL = 'ASHR'  # CSI 300 Index ETF
NAV_SYMBOL = '000300.SS'  # NAV proxy, such as the index it tracks
THRESHOLD_VOLUME = 100000  # Example threshold for significant volume
THRESHOLD_PREMIUM_DISCOUNT = 0.01  # Example threshold for premium/discount (1%)
EMAIL_SENDER = 'your_email@example.com'
EMAIL_PASSWORD = 'your_password'
EMAIL_RECEIVER = 'team_dg@example.com'
SMTP_SERVER = 'smtp.example.com'
SMTP_PORT = 587

def fetch_etf_data(symbol, period='1d', interval='1m'):
    etf_data = yf.download(tickers=symbol, period=period, interval=interval)
    etf_data.reset_index(inplace=True)
    etf_data.rename(columns={'Datetime': 'timestamp'}, inplace=True)
    return etf_data

def fetch_nav_data(symbol, period='1d', interval='1m'):
    nav_data = yf.download(tickers=symbol, period=period, interval=interval)
    nav_data.reset_index(inplace=True)
    nav_data.rename(columns={'Datetime': 'timestamp', 'Close': 'nav'}, inplace=True)
    return nav_data[['timestamp', 'nav']]

def compute_premium_discount(etf_data, nav_data):
    df = pd.merge(etf_data, nav_data, on='timestamp', how='inner')
    df['premium_discount'] = (df['Close'] - df['nav']) / df['nav']
    return df

def detect_significant_flows(df, volume_threshold, premium_discount_threshold):
    significant_flows = df[(df['Volume'] > volume_threshold) | 
                           (np.abs(df['premium_discount']) > premium_discount_threshold)]
    return significant_flows

def send_email(subject, body, sender, receiver, smtp_server, smtp_port, password):
    msg = MIMEMultipart()
    msg['From'] = sender
    msg['To'] = receiver
    msg['Subject'] = subject
    
    msg.attach(MIMEText(body, 'plain'))
    
    server = smtplib.SMTP(smtp_server, smtp_port)
    server.starttls()
    server.login(sender, password)
    text = msg.as_string()
    server.sendmail(sender, receiver, text)
    server.quit()

def main():
    etf_data = fetch_etf_data(ETF_SYMBOL)
    nav_data = fetch_nav_data(NAV_SYMBOL)
    
    if etf_data.empty or nav_data.empty:
        print("Failed to fetch data.")
        return
    
    df = compute_premium_discount(etf_data, nav_data)
    
    # Filter data for the last minute
    last_minute_data = df[df['timestamp'] > datetime.now() - timedelta(minutes=1)]
    
    significant_flows = detect_significant_flows(last_minute_data, THRESHOLD_VOLUME, THRESHOLD_PREMIUM_DISCOUNT)
    
    if not significant_flows.empty:
        subject = "Significant ETF Flows Detected"
        body = f"Significant flows/volumes detected on ETFs:\n\n{significant_flows.to_string(index=False)}"
        send_email(subject, body, EMAIL_SENDER, EMAIL_RECEIVER, SMTP_SERVER, SMTP_PORT, EMAIL_PASSWORD)
        print("Email sent.")
    else:
        print("No significant flows detected.")

if __name__ == "__main__":
    main()
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Configuration
ETF_SYMBOL = 'ASHR'  # CSI 300 Index ETF
NAV_SYMBOL = '000300.SS'  # NAV proxy, such as the index it tracks
THRESHOLD_VOLUME = 100000  # Example threshold for significant volume
THRESHOLD_PREMIUM_DISCOUNT = 0.01  # Example threshold for premium/discount (1%)

def fetch_etf_data(symbol, period='1d', interval='1m'):
    etf_data = yf.download(tickers=symbol, period=period, interval=interval)
    etf_data.reset_index(inplace=True)
    etf_data.rename(columns={'Datetime': 'timestamp'}, inplace=True)
    return etf_data

def fetch_nav_data(symbol, period='1d', interval='1m'):
    nav_data = yf.download(tickers=symbol, period=period, interval=interval)
    nav_data.reset_index(inplace=True)
    nav_data.rename(columns={'Datetime': 'timestamp', 'Close': 'nav'}, inplace=True)
    return nav_data[['timestamp', 'nav']]

def compute_premium_discount(etf_data, nav_data):
    df = pd.merge(etf_data, nav_data, on='timestamp', how='inner')
    df['premium_discount'] = (df['Close'] - df['nav']) / df['nav']
    return df

def detect_significant_flows(df, volume_threshold, premium_discount_threshold):
    significant_flows = df[(df['Volume'] > volume_threshold) | 
                           (np.abs(df['premium_discount']) > premium_discount_threshold)]
    return significant_flows

def plot_signals(df, significant_flows):
    plt.figure(figsize=(14, 8))

    # Plot ETF close price
    plt.plot(df['timestamp'], df['Close'], label='ETF Close Price', color='blue')
    
    # Highlight significant volume
    plt.scatter(significant_flows['timestamp'], significant_flows['Close'], color='red', label='Significant Volume/Premium', marker='o')

    # Plot premium/discount as a secondary y-axis
    plt.twinx()
    plt.plot(df['timestamp'], df['premium_discount'], label='Premium/Discount', color='green', alpha=0.5)

    plt.title('ETF Price and Premium/Discount with Significant Flow Signals')
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.legend(loc='upper left')
    plt.show()

def main():
    etf_data = fetch_etf_data(ETF_SYMBOL)
    nav_data = fetch_nav_data(NAV_SYMBOL)
    
    if etf_data.empty or nav_data.empty:
        print("Failed to fetch data.")
        return
    
    df = compute_premium_discount(etf_data, nav_data)
    
    # Filter data for the last minute
    last_minute_data = df[df['timestamp'] > datetime.now() - timedelta(minutes=1)]
    
    significant_flows = detect_significant_flows(last_minute_data, THRESHOLD_VOLUME, THRESHOLD_PREMIUM_DISCOUNT)
    
    plot_signals(df, significant_flows)

if __name__ == "__main__":
    main()
```
