```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.vector_ar.vecm import coint_johansen
from statsmodels.tsa.stattools import coint
from itertools import combinations
import statsmodels.api as sm

# Define the tickers for demonstration (excluding S&P 500 index)
tickers = ["MSFT", "NVDA", "AAPL", "AMZN", "META", "GOOGL", "AVGO", "LLY", "JPM", "XOM", "TSLA", "UNH", "V", "PG", "COST", "MA", "JNJ", "HD"]

start_date = '2013-01-01'
end_date = '2024-06-01'

# Download the data
data = yf.download(tickers, start=start_date, end=end_date)['Adj Close']
data.dropna(inplace=True)

# Download S&P 500 index data
sp500 = yf.download('^GSPC', start=start_date, end=end_date)['Adj Close']
sp500.dropna(inplace=True)

# Calculate log returns
log_returns = np.log(data).diff().dropna()
sp500_log_returns = np.log(sp500).diff().dropna()

# Normalize log returns
normalized_log_returns = (log_returns - log_returns.mean()) / log_returns.std()

# Function to test cointegration for a pair using Johansen test
def johansen_cointegration_test(pair, data):
    log_ret_pair = data[list(pair)]
    johansen_test = coint_johansen(log_ret_pair, det_order=0, k_ar_diff=1)
    trace_statistic = johansen_test.lr1
    critical_values = johansen_test.cvt[:, 1]  # 5% critical values
    num_cointegrating_relations = sum(trace_statistic > critical_values)
    cointegration_vectors = johansen_test.evec if num_cointegrating_relations > 0 else None
    return num_cointegrating_relations, pair, cointegration_vectors

# Function to test cointegration for a pair using Engle-Granger test
def engle_granger_cointegration_test(pair, data):
    log_ret_pair = data[list(pair)]
    score, pvalue, _ = coint(log_ret_pair.iloc[:, 0], log_ret_pair.iloc[:, 1])
    return pvalue, pair

# Split the data into in-sample (first 6 months) and out-of-sample
split_date = '2020-07-01'
first_six_months_end = pd.to_datetime(split_date) + pd.DateOffset(months=6)
in_sample_data = normalized_log_returns[:split_date]
first_six_months_data = normalized_log_returns[:first_six_months_end]
out_sample_data = normalized_log_returns[split_date:]

# Get all possible pairs
pairs = list(combinations(in_sample_data.columns, 2))

# Test cointegration for all pairs using in-sample data
johansen_results = [johansen_cointegration_test(pair, in_sample_data) for pair in pairs]
engle_granger_results = [engle_granger_cointegration_test(pair, in_sample_data) for pair in pairs]

# Combine the pairs and remove duplicates
pairs_johansen = [result[1] for result in johansen_results if result[0] > 0]  # Only keep pairs with at least 1 cointegrating relation
pairs_engle_granger = [result[1] for result in engle_granger_results if result[0] < 0.05]  # Only keep pairs with p-value < 0.05
combined_pairs = list(set(pairs_johansen + pairs_engle_granger))

# Define the transaction cost rate
transaction_cost_rate = 0.001

# Parameters for entry, exit, and stop-loss
entry_threshold = 2
exit_threshold = 0.5
stop_loss_threshold = 3

# Calculate the spreads, Z-scores, positions, and PnL with transaction costs for the selected pairs
def calculate_spreads_positions_pnl_with_costs(pair, in_sample_data, out_sample_data):
    spread = in_sample_data[pair[0]] - in_sample_data[pair[1]]
    spread_out_sample = out_sample_data[pair[0]] - out_sample_data[pair[1]]
    spread_mean = spread_out_sample.rolling(window=252).mean()
    spread_std = spread_out_sample.rolling(window=252).std()
    z_score = (spread_out_sample - spread_mean) / spread_std

    positions = []
    for i in range(len(z_score)):
        if z_score[i] > entry_threshold:
            positions.append([-1, 1])  # Short first, long second
        elif z_score[i] < -entry_threshold:
            positions.append([1, -1])  # Long first, short second
        elif abs(z_score[i]) < exit_threshold:
            positions.append([0, 0])  # Exit positions
        else:
            positions.append(positions[-1] if positions else [0, 0])

    positions_df = pd.DataFrame(positions, index=z_score.index, columns=[pair[0], pair[1]])
    
    # Calculate daily PnL
    daily_pnl = (positions_df.shift(1) * log_returns.loc[out_sample_data.index][[pair[0], pair[1]]]).sum(axis=1)
    
    # Calculate transaction costs
    transaction_costs = positions_df.diff().abs().sum(axis=1) * transaction_cost_rate
    transaction_costs.iloc[0] = 0  # No cost on the first day as there's no prior position

    # Subtract transaction costs from daily PnL
    daily_pnl -= transaction_costs

    return spread, z_score, positions_df, daily_pnl

# Calculate Sharpe ratio
def calculate_sharpe_ratio(daily_pnl):
    total_return = daily_pnl.sum()
    annualized_return = total_return * (252 / len(daily_pnl))  # Assuming 252 trading days in a year
    annualized_volatility = daily_pnl.std() * np.sqrt(252)
    sharpe_ratio = annualized_return / annualized_volatility
    return sharpe_ratio

# Calculate maximum drawdown
def calculate_max_drawdown(cumulative_pnl):
    drawdown = cumulative_pnl - cumulative_pnl.cummax()
    max_drawdown = drawdown.min()
    return max_drawdown

# Calculate Calmar ratio
def calculate_calmar_ratio(annualized_return, max_drawdown):
    calmar_ratio = annualized_return / abs(max_drawdown)
    return calmar_ratio

# Evaluate all combined pairs and select top 8 by Sharpe ratio using the first 6 months data
sharpe_ratios = []
daily_pnls = []
for pair in combined_pairs:
    spread, z_score, positions_df, daily_pnl = calculate_spreads_positions_pnl_with_costs(pair, first_six_months_data, out_sample_data)
    sharpe_ratio = calculate_sharpe_ratio(daily_pnl)
    sharpe_ratios.append((sharpe_ratio, pair, daily_pnl))
    daily_pnls.append(daily_pnl)

# Sort by Sharpe ratio and select top 8 pairs
sharpe_ratios.sort(reverse=True, key=lambda x: x[0])
top_8_pairs = sharpe_ratios[:8]

# Combine top 8 pairs into a single portfolio with $1000 initial investment for the whole portfolio
initial_investment = 1000

# Calculate the proportion of the investment for each pair
investment_per_pair = initial_investment / 8

# Calculate the daily PnL for the portfolio
portfolio_daily_pnl = sum([pnl * investment_per_pair for _, _, pnl in top_8_pairs])

# Calculate the cumulative PnL for the portfolio
portfolio_cumulative_pnl = portfolio_daily_pnl.cumsum()

# Calculate the cumulative value of the portfolio
portfolio_value = initial_investment + portfolio_cumulative_pnl

# Calculate performance metrics
final_portfolio_value = portfolio_value[-1]
annualized_return = (final_portfolio_value / initial_investment) ** (252 / len(portfolio_value)) - 1
max_drawdown = calculate_max_drawdown(portfolio_cumulative_pnl)
calmar_ratio = calculate_calmar_ratio(annualized_return, max_drawdown)
sharpe_ratio = calculate_sharpe_ratio(portfolio_daily_pnl)

# Set the start date for plotting
plot_start_date = '2021-05-01'

# Plot individual performances starting from the specified date
plt.figure(figsize=(12, 8))
for sharpe, pair, daily_pnl in top_8_pairs:
    cumulative_pnl = daily_pnl.cumsum()
    plt.plot(cumulative_pnl[plot_start_date:], label=f'{pair} (Sharpe: {sharpe:.2f})')
plt.title('Cumulative PnL of Top 8 Pairs by Sharpe Ratio (Out-of-Sample)')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Cumulative PnL')
plt.show()

# Plot portfolio performance starting from the specified date
plt.figure(figsize=(12, 8))
plt.plot(portfolio_value[plot_start_date:], label='Portfolio Value (Initial Investment $1000)')
plt.title('Cumulative Value of the Portfolio (Top 8 Pairs by Sharpe Ratio)')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Portfolio Value')
plt.show()

# Plot daily returns of the portfolio
plt.figure(figsize=(12, 8))
plt.plot(portfolio_daily_pnl[plot_start_date:], label='Portfolio Daily Returns')
plt.title('Daily Returns of the Portfolio (Top 8 Pairs by Sharpe Ratio)')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Daily Returns')
plt.show()

# Calculate rolling Sharpe ratio for the portfolio
rolling_sharpe = portfolio_daily_pnl.rolling(window=252).mean() / portfolio_daily_pnl.rolling(window=252).std() * np.sqrt(252)

# Plot rolling Sharpe ratio of the portfolio
plt.figure(figsize=(12, 8))
plt.plot(rolling_sharpe[plot_start_date:], label='Portfolio Rolling Sharpe Ratio')
plt.title('Rolling Sharpe Ratio of the Portfolio (Top 8 Pairs by Sharpe Ratio)')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Sharpe Ratio')
plt.show()

# Print final portfolio value and performance metrics
print(f"Final Portfolio Value: {final_portfolio_value:.2f}")
print(f"Annualized Return: {annualized_return:.2%}")
print(f"Maximum Drawdown: {max_drawdown:.2f}")
print(f"Calmar Ratio: {calmar_ratio:.2f}")
print(f"Sharpe Ratio: {sharpe_ratio:.2f}")
```

### Detailed Explanation of Beta, Volatility, and Dollar Neutrality in Pairs Trading

Pairs trading involves creating a market-neutral strategy that seeks to profit from the relative price movements of two related securities. To achieve this neutrality, we use various methods: beta neutrality, volatility neutrality, and dollar neutrality. Below, we explain the rationale and methods behind each approach.

#### 1. Beta Neutrality

**Why Beta Neutrality?**
- **Market Risk Mitigation**: Beta neutrality aims to eliminate the systematic risk associated with market movements. By neutralizing the beta, the portfolio's exposure to market-wide risk (i.e., movements in the overall market) is minimized.
- **Focus on Relative Value**: By hedging against market movements, the strategy focuses purely on the relative performance between the pair, which is the core idea behind pairs trading.

**How Beta Neutrality Works:**
- **Calculate Beta**: The beta of a stock measures its sensitivity to market movements. A beta of 1 means the stock moves in line with the market, while a beta greater than 1 indicates higher sensitivity, and a beta less than 1 indicates lower sensitivity.
- **Hedge Ratio**: To achieve beta neutrality, calculate the hedge ratio based on the betas of the two stocks. For example, if Stock A has a beta of 1.2 and Stock B has a beta of 0.8, the hedge ratio would be determined such that the overall beta of the long and short positions combined is zero.
- **Position Sizing**: Adjust the position sizes in the two stocks to neutralize the beta. This involves taking larger positions in lower-beta stocks and smaller positions in higher-beta stocks.

#### 2. Volatility Neutrality

**Why Volatility Neutrality?**
- **Risk Management**: Volatility neutrality aims to balance the risk exposure between the two stocks. High-volatility stocks are more unpredictable, and without balancing, they can dominate the portfolio's risk profile.
- **Stable Returns**: By equalizing the volatility, the strategy seeks to ensure that both legs of the trade contribute equally to the portfolio's risk and return, leading to more stable performance.

**How Volatility Neutrality Works:**
- **Measure Volatility**: Calculate the historical volatility of each stock, typically using standard deviation of returns over a specific period.
- **Position Sizing**: Adjust the position sizes such that the dollar amount invested in each stock is inversely proportional to its volatility. For example, if Stock A is twice as volatile as Stock B, the investment in Stock A would be half that of Stock B.
- **Hedge Ratio**: This ratio ensures that each stock contributes equally to the portfolio's overall volatility, balancing the risk exposure.

#### 3. Dollar Neutrality

**Why Dollar Neutrality?**
- **Simplified Risk Management**: Dollar neutrality aims to simplify the risk management process by ensuring that the dollar value of long positions equals the dollar value of short positions.
- **Capital Allocation**: This method focuses on the allocation of capital without directly considering beta or volatility, making it straightforward to implement.

**How Dollar Neutrality Works:**
- **Equal Dollar Amounts**: Allocate an equal dollar amount to both the long and short positions. For example, if you invest $10,000 in Stock A (long), you also short $10,000 worth of Stock B.
- **Rebalancing**: Regularly rebalance the positions to maintain dollar neutrality, as the market values of the stocks change over time.
- **Simple Hedge**: This method provides a simple hedge against market movements but does not account for differences in beta or volatility, which could lead to residual risks.

### Practical Considerations

**Choosing the Neutrality Approach:**
- **Market Conditions**: The choice of neutrality method can depend on current market conditions. For instance, beta neutrality is crucial in volatile markets to hedge against systemic risks.
- **Strategy Focus**: If the strategy focuses on capital allocation efficiency, dollar neutrality might be preferred. For risk-focused strategies, volatility neutrality could be more suitable.
- **Implementation Complexity**: Volatility and beta neutrality require more complex calculations and ongoing adjustments compared to dollar neutrality.

**Regular Monitoring and Rebalancing:**
- All methods require regular monitoring and rebalancing to maintain the desired neutrality. Market movements, changes in volatility, and shifts in beta can alter the initial balance, necessitating adjustments.

### Conclusion

Neutralizing market exposure through beta, volatility, and dollar neutrality is crucial in pairs trading to focus on the relative performance of paired stocks. Each method offers distinct advantages and challenges, and the choice depends on the strategy's goals, market conditions, and risk management preferences. Implementing these methods requires careful calculation, regular monitoring, and rebalancing to ensure the strategy remains effective and aligned with its objectives.
