```python
import pandas as pd

# Load the data from the Excel file
#df = pd.read_excel('data.xlsx')

# Sample data
data = {
    'Date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02'],
    'Strategy': ['Client1', 'Client1', 'Client2', 'Client2'],
    'Long/Short': ['Long', 'Short', 'Long', 'Short'],
    'Benchmark': ['SOF', 'SOF', 'HIBOR 1M', 'HIBOR 1M'],
    'Currency': ['USD', 'USD', 'HKD', 'HKD'],
    'Total Position': [1000000, 500000, 2000000, 1500000],
    'Funding Cost': [0.02, 0.01, 0.015, 0.02]
}

df = pd.DataFrame(data)

# Define benchmark rates (example values)
benchmark_rates = {
    'SOF': 0.01,        # 1% for SOF
    'HIBOR 1M': 0.015   # 1.5% for HIBOR 1M
}

# Calculate the cost of funding vs benchmark
def calculate_cost_of_funding(row):
    benchmark_rate = benchmark_rates[row['Benchmark']]
    funding_cost_annual = row['Funding Cost'] * 365   # Convert daily funding cost to annual
    cost_of_funding_bps = (funding_cost_annual - benchmark_rate) * 10000  # Convert to basis points
    return cost_of_funding_bps

df['Cost of Funding (bps)'] = df.apply(calculate_cost_of_funding, axis=1)

# Aggregate at the Strategy level
result = df.groupby('Strategy').agg({
    'Cost of Funding (bps)': 'mean',
    'Total Position': 'sum'
}).reset_index()

# Display the result
print(result)
```
