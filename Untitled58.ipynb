```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.vector_ar.vecm import coint_johansen
from statsmodels.tsa.stattools import coint
from itertools import combinations
import statsmodels.api as sm

#tech_tickers = ["MSFT", "GOOGL", "AAPL", "AMD", "AKAM", "TSLA", "AMZN"]
tech_tickers = ["AAPL", "MSFT", "NVDA", "GOOGL", "META"]
health_tickers = ["JNJ", "UNH", "PFE", "ABT", "LLY"]
finance_tickers = ["JPM", "BAC", "WFC", "C", "GS"]
consumer_tickers = ["AMZN", "TSLA", "HD", "NKE", "MCD"]

selected_tickers = tech_tickers + health_tickers + finance_tickers + consumer_tickers

start_date = '2013-01-01'
end_date = '2024-06-01'

# Download the data
data = yf.download(selected_tickers, start=start_date, end=end_date)['Adj Close']
data.dropna(inplace=True)

# Download S&P 500 index data
sp500 = yf.download('^GSPC', start=start_date, end=end_date)['Adj Close']
sp500.dropna(inplace=True)

# Calculate log returns
log_returns = np.log(data).diff().dropna()
sp500_log_returns = np.log(sp500).diff().dropna()

# Normalize log returns
normalized_log_returns = (log_returns - log_returns.mean()) / log_returns.std()

# Function to test cointegration for a pair using Johansen test
def johansen_cointegration_test(pair, data):
    log_ret_pair = data[list(pair)]
    johansen_test = coint_johansen(log_ret_pair, det_order=0, k_ar_diff=1)
    trace_statistic = johansen_test.lr1
    critical_values = johansen_test.cvt[:, 1]  # 5% critical values
    num_cointegrating_relations = sum(trace_statistic > critical_values)
    cointegration_vectors = johansen_test.evec if num_cointegrating_relations > 0 else None
    return num_cointegrating_relations, pair, cointegration_vectors

# Function to test cointegration for a pair using Engle-Granger test
def engle_granger_cointegration_test(pair, data):
    log_ret_pair = data[list(pair)]
    score, pvalue, _ = coint(log_ret_pair.iloc[:, 0], log_ret_pair.iloc[:, 1])
    return pvalue, pair

# Split the data into in-sample (first 6 months) and out-of-sample
split_date = '2020-07-01'
first_six_months_end = pd.to_datetime(split_date) + pd.DateOffset(months=6)
in_sample_data = normalized_log_returns[:split_date]
first_six_months_data = normalized_log_returns[:first_six_months_end]
out_sample_data = normalized_log_returns[split_date:]

# Generate pairs from the same sector
tech_pairs = list(combinations(tech_tickers, 2))
health_pairs = list(combinations(health_tickers, 2))
finance_pairs = list(combinations(finance_tickers, 2))
sector_pairs = tech_pairs + health_pairs + finance_pairs + list(combinations(consumer_tickers, 2))

# Test cointegration for all pairs using in-sample data
johansen_results = [johansen_cointegration_test(pair, in_sample_data) for pair in sector_pairs]
engle_granger_results = [engle_granger_cointegration_test(pair, in_sample_data) for pair in sector_pairs]

# Combine the pairs and remove duplicates
pairs_johansen = [result[1] for result in johansen_results if result[0] > 0]  # Only keep pairs with at least 1 cointegrating relation
pairs_engle_granger = [result[1] for result in engle_granger_results if result[0] < 0.05]  # Only keep pairs with p-value < 0.05
combined_pairs = list(set(pairs_johansen + pairs_engle_granger))

# Function to calculate beta relative to a benchmark
def calculate_beta(stock_returns, benchmark_returns):
    aligned_stock_returns, aligned_benchmark_returns = stock_returns.align(benchmark_returns, join='inner')
    X = sm.add_constant(aligned_stock_returns)
    model = sm.OLS(aligned_benchmark_returns, X).fit()
    return model.params[1]  # Beta coefficient

# Function to calculate position sizes for volatility, beta, and dollar neutrality
def calculate_neutral_positions(pair, positions_df, method):
    if method == 'volatility':
        volatility_stock1 = positions_df[pair[0]].std()
        volatility_stock2 = positions_df[pair[1]].std()
        position_ratio_stock1 = 1 / volatility_stock1
        position_ratio_stock2 = 1 / volatility_stock2
    elif method == 'beta':
        beta_stock1 = calculate_beta(log_returns[pair[0]], sp500_log_returns)
        beta_stock2 = calculate_beta(log_returns[pair[1]], sp500_log_returns)
        position_ratio_stock1 = 1 / beta_stock1
        position_ratio_stock2 = 1 / beta_stock2
    else:  # dollar neutrality
        position_ratio_stock1 = 1
        position_ratio_stock2 = 1
    
    total_ratio = position_ratio_stock1 + position_ratio_stock2
    position_ratio_stock1 /= total_ratio
    position_ratio_stock2 /= total_ratio
    
    return position_ratio_stock1, position_ratio_stock2

# Parameters for entry, exit, and stop-loss
entry_threshold = 2
exit_threshold = 0.5
stop_loss_threshold = 3

# Calculate the spreads, Z-scores, and positions for the selected pairs
def calculate_spreads_and_positions(pair, in_sample_data, out_sample_data):
    spread = in_sample_data[pair[0]] - in_sample_data[pair[1]]
    spread_out_sample = out_sample_data[pair[0]] - out_sample_data[pair[1]]
    spread_mean = spread_out_sample.rolling(window=252).mean()
    spread_std = spread_out_sample.rolling(window=252).std()
    z_score = (spread_out_sample - spread_mean) / spread_std

    positions = []
    for i in range(len(z_score)):
        if z_score[i] > entry_threshold:
            positions.append([-1, 1])  # Short first, long second
        elif z_score[i] < -entry_threshold:
            positions.append([1, -1])  # Long first, short second
        elif abs(z_score[i]) < exit_threshold:
            positions.append([0, 0])  # Exit positions
        else:
            positions.append(positions[-1] if positions else [0, 0])
    
    positions_df = pd.DataFrame(positions, index=z_score.index, columns=[pair[0], pair[1]])
    return spread, z_score, positions_df

# Calculate Sharpe ratio
def calculate_sharpe_ratio(daily_pnl):
    total_return = daily_pnl.sum()
    annualized_return = total_return * (252 / len(daily_pnl))  # Assuming 252 trading days in a year
    annualized_volatility = daily_pnl.std() * np.sqrt(252)
    sharpe_ratio = annualized_return / annualized_volatility
    return sharpe_ratio

# Calculate the portfolio Sharpe ratio and annualized return
def calculate_portfolio_metrics(portfolio_daily_pnl):
    total_return = portfolio_daily_pnl.sum()
    annualized_return = total_return * (252 / len(portfolio_daily_pnl))  # Assuming 252 trading days in a year
    annualized_volatility = portfolio_daily_pnl.std() * np.sqrt(252)
    sharpe_ratio = annualized_return / annualized_volatility
    return sharpe_ratio, annualized_return

# Evaluate all combined pairs and select top 8 by Sharpe ratio using the first 6 months data
sharpe_ratios = []
daily_pnls = []
for pair in combined_pairs:
    spread, z_score, positions_df = calculate_spreads_and_positions(pair, first_six_months_data, out_sample_data)
    daily_pnl = (positions_df.shift(1) * log_returns.loc[first_six_months_end:][[pair[0], pair[1]]]).sum(axis=1)
    sharpe_ratio = calculate_sharpe_ratio(daily_pnl)
    sharpe_ratios.append((sharpe_ratio, pair, daily_pnl))
    daily_pnls.append(daily_pnl)

# Sort by Sharpe ratio and select top 8 pairs
sharpe_ratios.sort(reverse=True, key=lambda x: x[0])
top_8_pairs = sharpe_ratios[:8]

# Set the start date for plotting
plot_start_date = '2021-05-01'

# Plot individual performances starting from the specified date
plt.figure(figsize=(12, 8))
for sharpe, pair, daily_pnl in top_8_pairs:
    cumulative_pnl = daily_pnl.cumsum()
    plt.plot(cumulative_pnl[plot_start_date:], label=f'{pair} (Sharpe: {sharpe:.2f})')
plt.title('Cumulative PnL of Top 8 Pairs by Sharpe Ratio (Out-of-Sample)')
plt.legend()
plt.show()

# Combine top 8 pairs into a single portfolio with $1000 initial investment for the whole portfolio
initial_investment = 1000

# Calculate the proportion of the investment for each pair
investment_per_pair = initial_investment / 8

# Calculate daily PnL for the portfolio with different neutrality methods
portfolio_daily_pnls = {'volatility': [], 'beta': [], 'dollar': []}

for method in ['volatility', 'beta', 'dollar']:
    daily_pnls_neutral = []
    for _, pair, daily_pnl in top_8_pairs:
        _, _, positions_df = calculate_spreads_and_positions(pair, first_six_months_data, out_sample_data)
        position_ratio_stock1, position_ratio_stock2 = calculate_neutral_positions(pair, positions_df, method)
        adjusted_pnl = (positions_df.shift(1) * log_returns.loc[out_sample_data.index][[pair[0], pair[1]]] * [position_ratio_stock1, position_ratio_stock2]).sum(axis=1)
        daily_pnls_neutral.append(adjusted_pnl * investment_per_pair)
    
    portfolio_daily_pnls[method] = sum(daily_pnls_neutral)

# Calculate cumulative PnL for each method
portfolio_cumulative_pnls = {method: pnl.cumsum() for method, pnl in portfolio_daily_pnls.items()}

# Calculate the cumulative value of the portfolio
portfolio_values = {method: initial_investment + cumulative_pnl for method, cumulative_pnl in portfolio_cumulative_pnls.items()}

# Plot portfolio performance starting from the specified date
for method, portfolio_value in portfolio_values.items():
    plt.figure(figsize=(12, 8))
    plt.plot(portfolio_value[plot_start_date:], label=f'Portfolio Value ({method.capitalize()} Neutrality)')
    plt.title(f'Cumulative Value of the Portfolio ({method.capitalize()} Neutrality)')
    plt.legend()
    plt.xlabel('Date')
    plt.ylabel('Portfolio Value')
    plt.show()

# Print final portfolio value for each method
for method, portfolio_value in portfolio_values.items():
    final_portfolio_value = portfolio_value[-1]
    print(f"Final Portfolio Value ({method.capitalize()} Neutrality): {final_portfolio_value:.2f}")

# Calculate and print portfolio metrics for the dollar neutrality method
sharpe_ratio, annualized_return = calculate_portfolio_metrics(portfolio_daily_pnls['dollar'])

print(f"Portfolio Sharpe Ratio: {sharpe_ratio:.2f}")

```

```python
# Calculate maximum drawdown
def calculate_max_drawdown(portfolio_value):
    cumulative_max = portfolio_value.cummax()
    drawdown = (portfolio_value - cumulative_max) / cumulative_max
    max_drawdown = drawdown.min()
    return max_drawdown

# Example calculation and print statement for the portfolio using Beta Neutrality
portfolio_value_beta = portfolio_values['beta']
max_drawdown_beta = calculate_max_drawdown(portfolio_value_beta)
print(f"Maximum Drawdown: {max_drawdown_beta:.2%}")
```
